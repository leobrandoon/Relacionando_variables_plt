```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Carga y manipulación

```python
df = pd.read_csv(r'C:\Users\leogo\Desktop\Desarrollos\Analitica\Curso_Ds_Udemy\Datasets+Data+Science+con+Python+-+Pandas\1.3\Info_pais.csv', encoding = 'ISO-8859-1',delimiter =';')
```

```python
df.columns
```

```python
df_order = df.sort_values("Esperanza de vida",ascending=True)
```

```python
df_order.head()
```

# Visualizaciones

```python
plt.scatter(df_order["Renta per capita"],df_order["Esperanza de vida"])
#asiganos leyendas a la grafica
plt.title("Renta percapita vs Esperanza de vida")
plt.xlabel("Renta percapita") 
plt.ylabel("Esperanza de vida")
```

```python
df_order["pob_normal"]=df_order["Poblacion"]/(max(df_order["Poblacion"])/10000)
```

```python
df_order.head()
```

```python
#Ajustando grafico por tamaño segun columna normal o ponderada
plt.scatter(df_order["Renta per capita"],df_order["Esperanza de vida"],s=df_order["pob_normal"],c=df_order["pob_normal"])
#asiganos leyendas a la grafica
plt.title("Renta percapita vs Esperanza de vida")
plt.xlabel("Renta percapita") 
plt.ylabel("Esperanza de vida")

#mejorando las pulgadas de visualizacion
fig=plt.gcf()
fig.set_size_inches(14.5, 10)

#recorremos la lista con un for para etiquetal los 10 paises mas grandes
for i in range(1,10):
    plt.annotate(df_order["País"][i],(df_order["Renta per capita"][i],df_order["Esperanza de vida"][i]))
    
#Mejoramos la visul de los ejes
plt.yticks(ticks=range(1,120,10)) #desde el 1 al 120 saltos de 10
```
